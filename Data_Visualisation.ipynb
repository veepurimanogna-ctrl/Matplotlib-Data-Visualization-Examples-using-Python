```python
import matplotlib.pyplot as plt

subjects = ['Python', 'Java', 'C++', 'SQL']
students = [45, 38, 30, 50]

plt.bar(subjects, students)

plt.title("Students Enrolled")
plt.xlabel("Subjects")
plt.ylabel("Number of Students")

plt.show()
```

```python

plt.bar(subjects, students, color='green')
plt.show()
```

```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [25, 30, 40, 35, 50]

plt.plot(months, sales)

plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")

plt.show()
```

```python
plt.plot(months, sales,
         marker='o',
         linestyle='--',
         color='red')
plt.show()
```

```python

plt.plot(months, sales,
         marker='o',
         linestyle='--',
         color='blue')
plt.show()
```

```python

import matplotlib.pyplot as plt

labels = ['Python', 'Java', 'C++', 'SQL']
students = [40, 30, 20, 10]

plt.pie(students,
        labels=labels,
        autopct='%1.1f%%',
        startangle=90)

plt.title("Students Learning Programming Languages")

plt.show()

```
